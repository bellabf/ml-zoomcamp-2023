```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
car_df=pd.read_csv("datasets/data_cars.csv")
columns_to_keep=["Make",
                 "Model",
                 "Year",
                 "Engine HP",
                 "Engine Cylinders",
                 "Transmission Type",
                 "Vehicle Style",
                 "highway MPG",
                 "city mpg",
                 "MSRP"]
```

```python
homework_data=car_df[columns_to_keep]
homework_data.columns = columns_to_keep.columns.str.replace(' ', '_').str.lower()
homework_data.fillna(0, inplace=True)
mean_price=np.mean(homework_data["MSRP"])
homework_data["binary_price"]= homework_data["MSRP"] > mean_price
```
